## Introduction to Probability in Python

```python
# Representing probabilities
prob_heads = 0.5  # Probability of getting heads on a fair coin

# Complementary events
prob_tails = 1 - prob_heads

print(f"Probability of heads: {prob_heads}")
print(f"Probability of tails: {prob_tails}")

# Multiple events
prob_double_heads = prob_heads * prob_heads
print(f"Probability of two heads in a row: {prob_double_heads}")
```

## Random Number Generation for simulating probabilistic events.

```python
import random

# Set a seed for reproducibility
random.seed(42)

# Generate a random number between 0 and 1
random_num = random.random()
print(f"Random number: {random_num}")

# Simulate a coin flip
coin_flip = "Heads" if random.random() < 0.5 else "Tails"
print(f"Coin flip result: {coin_flip}")

# Simulate rolling a die
die_roll = random.randint(1, 6)
print(f"Die roll result: {die_roll}")
```

## Probability Distribution 

```python
import numpy as np
import matplotlib.pyplot as plt

# Binomial distribution
n, p = 10, 0.5
binomial_dist = np.random.binomial(n, p, 1000)

plt.hist(binomial_dist, bins=range(12), align='left', rwidth=0.8)
plt.title("Binomial Distribution")
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.show()

# Normal distribution
mu, sigma = 0, 1
normal_dist = np.random.normal(mu, sigma, 1000)

plt.hist(normal_dist, bins=30)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Monte Carlo Simulation : a powerful tool for estimating probabilities.

```python
def estimate_pi(num_points):
    inside_circle = 0
    total_points = num_points
    
    for _ in range(total_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if x*x + y*y <= 1:
            inside_circle += 1
    
    pi_estimate = 4 * inside_circle / total_points
    return pi_estimate

# Estimate π
num_simulations = 1000000
estimated_pi = estimate_pi(num_simulations)
print(f"Estimated value of π: {estimated_pi}")
print(f"Actual value of π: {np.pi}")
```

## Bayes Theorem

```python
def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

# Example: Medical test
sensitivity = 0.99  # True positive rate
specificity = 0.95  # True negative rate
prevalence = 0.01   # Disease prevalence

# Probability of positive test given disease
p_pos_given_disease = sensitivity

# Probability of positive test given no disease
p_pos_given_no_disease = 1 - specificity

# Overall probability of a positive test
p_pos = p_pos_given_disease * prevalence + p_pos_given_no_disease * (1 - prevalence)

# Probability of disease given a positive test
p_disease_given_pos = bayes_theorem(prevalence, p_pos_given_disease, p_pos)

print(f"Probability of disease given a positive test: {p_disease_given_pos:.4f}")
```
