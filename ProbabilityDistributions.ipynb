## Binomial Distribution
- The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.

- Example: Number of heads in 10 coin flips.

```python

import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.5  # 10 trials, 50% probability of success
x = np.arange(0, n+1)
y = np.random.binomial(n, p, 1000)

plt.hist(y, bins=n+1, density=True)
plt.title("Binomial Distribution")
plt.show()
```

## Poisson Distribution
- The Poisson distribution models the number of events occurring in a fixed interval of time or space.
- Example: Number of customers arriving at a store in an hour.

```python
lambda_param = 5  # average number of events
x = np.arange(0, 20)
y = np.random.poisson(lambda_param, 1000)

plt.hist(y, bins=20, density=True)
plt.title("Poisson Distribution")
plt.show()
```

## Continuous Probability Distributions
- Continuous distributions deal with random variables that can take on any value within a given range.
## Normal Distribution
- The normal distribution, also known as Gaussian distribution, is symmetric and bell-shaped.
Example: Heights of adults in a population.

```python
mu, sigma = 170, 10  # mean and standard deviation
x = np.random.normal(mu, sigma, 1000)

plt.hist(x, bins=30, density=True)
plt.title("Normal Distribution")
plt.show()
```

## Exponential Distribution
- The exponential distribution models the time between events in a Poisson process.
- Example: Time between earthquakes.

```python
lambda_param = 0.5  # rate parameter
x = np.random.exponential(1/lambda_param, 1000)

plt.hist(x, bins=30, density=True)
plt.title("Exponential Distribution")
plt.show()
```

## Uniform Distribution
- In a uniform distribution, all outcomes are equally likely.
- Example: The number of times cars wait at a red light.

```python
a, b = 0, 10  # lower and upper bounds
x = np.random.uniform(a, b, 1000)

plt.hist(x, bins=30, density=True)
plt.title("Uniform Distribution")
plt.show()
```

## Log-Normal Distribution
- The log-normal distribution is used when the logarithm of the variable is normally distributed.
- Example: The average body weight of different mammal species.

```python
mu, sigma = 0, 0.5  # mean and standard deviation of log(X)
x = np.random.lognormal(mu, sigma, 1000)

plt.hist(x, bins=30, density=True)
plt.title("Log-Normal Distribution")
plt.show()
```
